```python
df["TotalCharges"] = df['TotalCharges'].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
def conv(value):
  if value == 1:
    return "Yes"
  else:
    return "No"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df
```

```python
ax = sn.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
mp.show()
```

```python
mp.figure(figsize = (4,4))
pc = df.groupby("Churn").agg({"Churn":"count"})
mp.pie(pc['Churn'], labels = pc.index,autopct = "%1.2f%%" )
mp.title("Total Customer By Churn")
mp.show()
```

From the given chart we can conclude that **26.54%** of customers have **Churn Out**


```python
sn.countplot(x  = "gender", data = df, hue = "Churn")
mp.title("Churn By Gender")
mp.show()
```

```python
mp.figure(figsize = (3,3))
sn.countplot(x = "SeniorCitizen", data = df,hue = "Churn")
mp.title(" Count by SeniorCitizen")
mp.show()
```

```python
crosstab = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Step 2: Plot a stacked bar chart
ax = crosstab.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], figsize=(5, 5))

# Step 3: Add labels to each section of the bar
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center', color='black')

mp.title("Churn Percentage by Senior Citizen")
mp.xlabel("Senior Citizen")
mp.ylabel("Percentage (%)")
mp.legend(title="Churn")
mp.show()
```

```python
mp.figure(figsize = (5,4))
ax = sn.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
mp.title(" Count by SeniorCitizen")
mp.show()
```

```python
mp.figure(figsize = (9,3))
sn.histplot(x = "tenure",data =df,bins = 70,hue = "Churn")
mp.show()
```

people who have use our services for a long time have stayed and people who have use our services for 1 or 2 month have churned

```python
mp.figure(figsize = (5,4))
ax = sn.countplot(x = "Contract", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
mp.title(" Count of Customer by  Contract")
mp.show()
```

people who have month-to-month contract are likely to churn then from those who have 1 or 2 years of contract

```python
df.columns.values
```

```python


# Assuming 'df' is your DataFrame

# List of columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the matplotlib figure with subplots in a grid layout
fig, axes = mp.subplots(nrows=3, ncols=3, figsize=(12, 11))
fig.suptitle("Count Plots for Service Features")

# Flatten axes for easy iteration
axes = axes.flatten()

# Loop through each column and create a countplot in the respective subplot
for i, col in enumerate(columns):
    sn.countplot(x=col, data=df, ax=axes[i], palette="Set2",hue = "Churn")
    axes[i].set_title(f"Count of {col}")
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Adjust layout for better readability
mp.tight_layout(rect=[0, 0.03, 1, 0.95])  # leave space for main title
mp.show()
```

**Customers lacking services like OnlineSecurity, OnlineBackup, DeviceProtection, and TechSupport show higher churn rates, indicating these features might help retain customers. Fiber optic internet users also exhibit higher churn, suggesting potential dissatisfaction. Overall, enhancing support and security-related services could be key to reducing churn.**
